# Notebook Intro:

In this notebook, I import cleaned training data and perform basic EDA on it.  This includes seeing the correlation between each column and the sale price (which will exclude non-numeric columns for now). 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import cleaned training data
filepath = '../datasets/interim_files/train_clean.csv'

df = pd.read_csv(filepath)
df.head()
```

```python
df.corr()[['SalePrice']]
```

```python
plt.figure(figsize=(7,15))

sns.heatmap(df.corr()[['SalePrice']].sort_values('SalePrice'),
           annot=True,
           vmin=-1,
           vmax=1,
           cmap='viridis');

plt.title('Sale Price vs all numeric features');
```

```python
# top 5 features in plot
Features = ['Overall Qual','Gr Liv Area','Garage Area','Garage Cars','Total Bsmt SF']

#check to see if related
df[Features].corr()

sns.heatmap(df[Features].corr(),
           annot=True,
           vmin=-1,
           vmax=1,
           cmap='viridis');

plt.title('Top 5 correlated features to Sale Price')
```

```python
# top 5 features in plot
Features = ['Overall Qual','Gr Liv Area','Garage Area','Garage Cars','Total Bsmt SF','SalePrice']

#check to see if related
df[Features].corr()

sns.heatmap(df[Features].corr()[['SalePrice']][0:5],
           annot=True,
           vmin=-1,
           vmax=1,
           cmap='viridis');
plt.title('Sale Price vs top 5 correlated features')
```

```python
plt.figure(figsize = (10,5))
plt.title('Overall Quality Rating vs Sale Price')
plt.bar(df['Overall Qual'],df['SalePrice']);
plt.xlabel('Quality Rating')
plt.ylabel('Sale Price [$]')
```
